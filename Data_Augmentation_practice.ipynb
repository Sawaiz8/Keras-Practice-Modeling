```python
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing import image #like opencv
from keras.preprocessing.image import ImageDataGenerator
```

```python
img  = keras.utils.load_img('/content/Elephant-Train (1).jpeg', target_size = (200,200))
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.8,
    #horizontal_flip = True,
    #width_shift_range = 0.2,
    #height_shift_range = 0.2,    
)
```

```python
!rm -rf AugmentedD
```

```python
img = keras.utils.img_to_array(img)
```

```python
input_batch = img.reshape(1,200,200,3) #Adding a new shape
```

```python
i = 0 
for output in datagen.flow(input_batch, batch_size = 1, save_to_dir = "/content/AugmentedD"):
  i += 1
  if i ==10:
    break
```
